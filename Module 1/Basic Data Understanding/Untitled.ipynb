### Plotting Mathematical Functions

```python
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as stats
from matplotlib import pyplot as plt
from math import pi
```

```python
x = np.arange(-pi,pi,0.1)
y = np.sin(x)
plt.plot(x,y,'ro') #other values include 'ro','bo','r',etc
plt.plot(x,np.cos(x))
plt.xlabel('pi')
plt.ylabel('sin(x) and cos(x)')
plt.legend(['sin','cos'])
plt.show()
```

### Plotting Line Plots

```python
x = [1,2,3]
y = [1,4,9]
z = [10,5,0]
plt.plot(x,y)
plt.plot(x,z)
plt.xlabel('x')
plt.ylabel('y and z')
plt.legend(['y','z'])
plt.show()
```
